## Problem Statement 1:Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    parked = 0  # keeps track of current parked vehicles
    # looping through each vehicle log
    for log in vehicle_logs:
        if log == "IN":
            parked += 1  # increase count when vehicle enters
        elif log == "OUT" and parked > 0:
            parked -= 1  # decrease count when vehicle exits
    print("Currently Parked Vehicles:", parked)
    # checking if parking exceeds capacity
    if parked > capacity:
        print("Parking Status: Over Capacity Alert!")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN","IN","IN","OUT","IN","IN","OUT"]
smart_parking_system(capacity, logs)
```

## Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    eta = distance * 5  # base time
    # traffic delay
    if traffic == "High":
        eta += 15
    elif traffic == "Medium":
        eta += 10
    # weather condition 
    if weather == "Stormy":
        eta += 20  # only stormy adds delay
    print("Estimated Delivery Time:", eta, "minutes")
estimate_delivery_time(8,"High","Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)  # total booked seats
    occupancy = (booked_count / total_seats) * 100  # percentage calculation
    print("Occupancy:", int(occupancy), "%")
    # checking show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
total = 200
booked = [1]*150
seat_occupancy(total, booked)
```

## Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    total = 0
    # calculating total cpu usage
    for value in cpu_readings:
        total += value
    average = total / len(cpu_readings)  # average calculation
    print("Average CPU Load:", int(average), "%")
    # classification based on average value
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")
cpu_data = [45,60,70,85,90]
classify_server_load(cpu_data)
```

## Problem 5: Smart Classroom Resource Usage Monitor

```python
def resource_monitor(resource_usage):
    overused = []  # list to store overused resources
    # checking usage hours
    for resource, hours in resource_usage.items():
        if hours > 8:  # threshold condition
            overused.append(resource)
    if overused:
        print("Overused Resources:", ",".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage = {"Projector":6,"AC":9,"Lights":4}
resource_monitor(usage)
```

## Problem Statement 6: Online Event Registration Capacity Controller

```python
def registration_controller(capacity, registrations):
    # checking if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
registration_controller(100,105)
```
