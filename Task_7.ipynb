## Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # default status
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(self.light_name + " is " + self.status)
# Taking input
name = input("Light Name: ")
action = input("Action (ON/OFF): ")
light1 = SmartLight(name)
if action.upper() == "ON":
    light1.turn_on()
else:
    light1.turn_off()
light1.display_status()
```

## Problem 2: Employee ID Card System

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)
# Taking input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")
emp1 = Employee(name, emp_id, dept)
emp1.display_id_card()
```

## Problem 3: Mobile Contact Record

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)
# Taking input
name = input("Contact Name: ")
phone = input("Phone Number: ")
contact1 = Contact(name, phone)
contact1.display_contact()
```

## Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    def display_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))   # converting price to string
# Taking input
name = input("Product Name: ")
price = int(input("Price: "))
product1 = Product(name, price)
product1.display_price_tag()
```

## Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")
# Taking input
name = input("Movie Name: ")
rating = float(input("Rating: "))
movie1 = Movie(name, rating)
movie1.display_movie()
```

## Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)
# Taking input
name = input("Customer Name: ")
address = input("Address: ")
delivery1 = Delivery(name, address)
delivery1.display_delivery()
```
