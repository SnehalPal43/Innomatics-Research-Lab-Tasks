```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    # Check minimum recharge amount
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    
    # Check if plan is valid
    if amount not in valid_plans:
        return "Invalid plan selected."
    
    return "Recharge successful!"

# Retry system using while loop
while True:
    recharge_amount = int(input("Enter recharge amount: "))
    result = validate_recharge(recharge_amount)
    print(result)
    
    if result == "Recharge successful!":
        break
    else:
        print("Please try again.\n")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert!")
        else:
            print(f"{item}: Stock OK")

# Sample inventory
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 8,
    "Monitor": 18
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"

# Example
student_marks = [60, 55, 70, 40, 50]
print(process_result(student_marks))
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    
    fare = base_fare + (distance * per_km)
    
    # Add 25% extra during peak hours
    if peak.lower() == "yes":
        fare += fare * 0.25
    
    return fare

# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak_hour)
    print(f"Total Fare: ₹{total_fare}")
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage}% - Eligible"
    else:
        return f"Attendance: {percentage}% - Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(check_attendance(attendance))
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8."
    elif not has_digit:
        return "Weak Password: Must contain at least one digit."
    elif not has_special:
        return "Weak Password: Must contain at least one special character (@#$)."
    else:
        return "Strong Password!"

# Example
user_password = input("Enter password: ")
print(check_password(user_password))
```
