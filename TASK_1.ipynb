```python
#Q1:
username=input("Enter Username:")
password=input("Enter Password:")
if username=='admin' and password=='1234': # Check if username and password match the predefined values
    print("Login Successful") # If both are correct
else:
    print("Invalid Credentials") # If either username or password is incorrect
```

```python
#Q2:
marks = [45, 78, 90, 33, 60]
# Initialize counters for pass and fail students
pass_count=0
fail_count=0
# Loop through each mark in the list
for mark in marks: 
    if mark >=50:
        pass_count+=1
    else:
        fail_count+=1
print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
    
```

```python
#Q3:
names = [" Alice ", "bob", " CHARLIE "]
# Empty list to store cleaned names
cleaned_names=[]
# Loop through each name in the list
for name in names:
# Remove extra spaces from start and end, and convert to lowercase
    clean_name=name.strip().lower()
# Add the clean name to the cleaned_names list
    cleaned_names.append(clean_name)
print("cleaned names are:",cleaned_names)
```

```python
#Q4:
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length=len(msg)
    print("message:",msg)
    print("length:",length)
    if length>10:
        print("message is longer then 10 character") # Flag long message
    else:
        print("short message")
    print()# Print a blank line for readability
```

```python
#Q5:
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
log_count=0 #count error entries
for log in logs:
    if log=="ERROR":
        log_count+=1
print("total error :",log_count) #print total error count
```
