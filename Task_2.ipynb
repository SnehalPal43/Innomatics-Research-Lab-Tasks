```python
#Problem 1: Unique Words in a Sentence
sentence=input("Enter Sentence:")
# Splitting sentence into words
words=sentence.split()
# Convert list into set to get unique words
unique_words=set(words)
length=len(unique_words)
print("Unique Word Count:",length)
print("Unique Words:",set(sorted(unique_words)))
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
# Finding employee with highest salary
highest_employee=max(employees,key=employees.get)
highest_salary=employees[highest_employee]
print("highest salary:",highest_employee,"-",highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
max_number=numbers[0]
min_number=numbers[0]

for number in numbers:
    if number>max_number:
        max_number=number
    if number<min_number:
        min_number=number
print("List:",numbers)
print("Max:",max_number)
print("Min:",min_number)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price>1000:
        count+=1
print("product above 1000:",count)
        
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
count=0
for present in attendance:
    if present=="P":
        count+=1
total_days=len(attendance)
percentage=(count/total_days)*100
print("Total Percentage:",percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
# Convert list into set to remove duplicates
unique_number=set(phone_numbers)
print("unique phone numbers:",unique_number)
```

```python
#Problem 7: Count Character Frequency
text="pythonp"
#empty dictionary to store character counts
frequency={}
for char in text:
    if char in frequency:
        frequency[char]+=1 # Increase its count by 1
    else:
        frequency[char]=1 # If not present, add it with count 1
print(frequency)
```
