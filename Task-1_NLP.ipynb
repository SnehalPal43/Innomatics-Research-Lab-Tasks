## Task 1: Conceptual Understanding

```python
# 1. Difference between "Love" and "love" in NLP
# Answer:
# In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
# "Love" may appear at the beginning of a sentence (proper capitalization),
# while "love" is used in general context.
# To avoid duplication and improve consistency, we usually convert text to lowercase.

# 2. What happens if stopwords are not removed?
# Answer:
# If stopwords are not removed, the dataset will contain many common words like "is", "the", "and".
# These words do not add much meaning and can reduce model performance
# by increasing noise and dimensionality.

# 3. Two real-world scenarios where removing stopwords can be harmful
# Answer:
# (a) Sentiment Analysis:
# Words like "not" are important. Removing them can change meaning.
# Example: "I am not happy" → "happy" (wrong meaning)
#
# (b) Question Answering Systems:
# Words like "who", "what", "where" are important for understanding intent.

# 4. Difference between stemming and lemmatization
# Answer:
# Stemming:
# It removes suffixes to get root form (may not be a valid word)
# Example: "running" → "run", "better" → "better" (sometimes incorrect)
#
# Lemmatization:
# It converts word to its base dictionary form using context
# Example: "running" → "run", "better" → "good"
```

## Task 2:Build Advanced Preprocessing Function

```python
import re
def preprocess_text(text):
    
    if not text or not isinstance(text, str):
        return [], ""
    
    # 1) Convert to lowercase
    text = text.lower()
    
    # 2) Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # 3) Remove email-like patterns
    text = re.sub(r"\S+@\S+", "", text)
    
    # 4) Remove numbers
    text = re.sub(r"\d+", "", text)
    
    # 5) Handle repeated characters (e.g., soooo → soo)
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    # 6) Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    
    # 7) Tokenize
    tokens = text.split()
    
    # Remove very short tokens (≤2), except 'no' and 'not'
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    
    # Join tokens to form cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)
    
    return cleaned_tokens, cleaned_sentence
```

```python
# Test sentences
test_sentences = [
    "I have 2 dogs",
    "This   is   good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "I am not happy"
]

# Apply preprocessing function
for s in test_sentences:
    tokens, cleaned = preprocess_text(s)
    print("Original:", s)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

## Task 3: Stress Testing

```python
# Sample sentences for stress testing
stress_sentences = [
    "Get 100% FREE access now!!!",                      # numbers + punctuation + mixed case
    "I absolutely looooved this product 😍😍",          # repeated characters + emojis
    "Worst service ever... 0/10",                       # numbers + punctuation
    "Call me at 9876543210",                            # numbers
    "This is THE best course!!!",                       # mixed case + punctuation
    "Visit https://openai.com now!",                    # URL
    "Nooooo this is baaad!!!",                          # repeated characters + punctuation
    "OK OK OK I got it",                                # repeated short tokens
    "Win $$$ now!!! Limited offer!!!",                  # symbols + punctuation
    "I am not happy with this"                          # includes 'not' exception
]

# Apply preprocessing function and display results
for sentence in stress_sentences:
    tokens, cleaned_sentence = preprocess_text(sentence)
    
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 60)
```

## Task 4: Token Analytics

```python
# Store results of preprocessing
analytics_results = []

for sentence in stress_sentences:
    tokens, cleaned_sentence = preprocess_text(sentence)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_token_length = round(sum(len(t) for t in tokens)/total_tokens, 2) if total_tokens > 0 else 0
    
    analytics_results.append({
        "original": sentence,
        "tokens": tokens,
        "cleaned_sentence": cleaned_sentence,
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": avg_token_length
    })

# Display analytics for each sentence
for result in analytics_results:
    print("Original Sentence:", result["original"])
    print("Total Tokens:", result["total_tokens"])
    print("Unique Tokens:", result["unique_tokens"])
    print("Average Token Length:", result["avg_token_length"])
    print("-"*60)

# Analysis Questions

# Which sentence had the most noise?
most_noise = max(analytics_results, key=lambda x: len(x["original"].split()) - x["total_tokens"])
print("Sentence with the most noise:")
print(most_noise["original"])
print("--------------------------------------------------")

# Which sentence retained the most meaningful tokens after cleaning?
most_meaningful = max(analytics_results, key=lambda x: x["total_tokens"])
print("Sentence that retained most meaningful tokens after cleaning:")
print(most_meaningful["original"])
```

## Task 5: Frequency Analysis

```python
from collections import Counter

# Combine all cleaned tokens from Task 3 / 4
all_tokens = []
for result in analytics_results:
    all_tokens.extend(result["tokens"])

# Count frequency of each token
token_counts = Counter(all_tokens)

# Top 10 most frequent words
top_10 = token_counts.most_common(10)
print("Top 10 Most Frequent Words:")
for word, freq in top_10:
    print(f"{word}: {freq}")

print("-"*50)

# Top 5 least frequent words
# Sort by frequency ascending
least_5 = sorted(token_counts.items(), key=lambda x: x[1])[:5]
print("Top 5 Least Frequent Words:")
for word, freq in least_5:
    print(f"{word}: {freq}")
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    # Lists to store all tokens and cleaned sentences
    all_tokens = []
    clean_sentences = []
    
    # Process each sentence
    for sentence in text_list:
        tokens, cleaned_sentence = preprocess_text(sentence)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned_sentence)
    
    # Return results in dictionary format
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
# Using the stress test sentences 
pipeline_result = full_pipeline(stress_sentences)

print("All Tokens Combined:", pipeline_result["tokens"])
print("-"*50)
print("Cleaned Sentences:", pipeline_result["clean_sentences"])
```
